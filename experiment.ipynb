```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder,StandardScaler,OneHotEncoder
```

```python
import pickle
```

```python
df = pd.read_csv('weatherAUS.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['RainTomorrow'])
```

```python
df.isna().sum()
```

```python
df['RainTomorrow'].value_counts()
```

```python
x = df.drop(columns=['Date','RainToday','RainTomorrow'],axis=1)
y = df['RainTomorrow']
```

```python
x.shape
```

```python
le = LabelEncoder()
```

```python
x['Location'] = le.fit_transform(x['Location'])
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
cat = x.select_dtypes(include='object').columns
num = x.select_dtypes(exclude='object').columns
```

```python
from sklearn.pipeline import Pipeline
```

```python
numerical_pipeline = Pipeline([
    ('imputer_numerical',SimpleImputer(strategy='median')),
    ('sclaer',StandardScaler())
])
catogorical_pipeline = Pipeline([
    ('catogorical_imputer',SimpleImputer(strategy='most_frequent')),
    ('one_hot_encoder',OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False))
])
```

```python
transformer = ColumnTransformer([
    ('numeric_transformer',numerical_pipeline,num),
    ('categorical_transformer',catogorical_pipeline,cat)
])
```

```python
x_train=transformer.fit_transform(x_train)
x_test = transformer.transform(x_test)
```

```python
with open('LabelEncoder.pkl','wb') as file:

    pickle.dump(le,file)
```

```python
with open('preprocessor.pkl','wb') as ohe:
    pickle.dump(transformer,ohe)
```

```python
x_train.dtype
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import datetime
```

```python
x_test.dtype
```

```python
models = Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(18,activation='relu'),
    Dense(8,activation='sigmoid'),
    Dense(1,activation='sigmoid')
])
```

```python
models.summary()
```

```python
models.compile(optimizer='Adam',loss='BinaryCrossentropy',metrics=['accuracy'])
```

```python
logdir = 'logs/fits/'+datetime.datetime.now().strftime('%Y-%M-%d-%H-%M_%S')
```

```python
tenser_flow_claaback = TensorBoard(logdir,histogram_freq=1)
```

```python
EarlyStopping_callback = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
lee = LabelEncoder()
y_train = lee.fit_transform(y_train)
y_test = lee.transform(y_test)
```

```python
weights = {0: 1.0, 1: 3.0}
```

```python
history = models.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,
                     callbacks=[tenser_flow_claaback,EarlyStopping_callback],class_weight=weights)
```

```python
models.save('newgen.h5')
```
